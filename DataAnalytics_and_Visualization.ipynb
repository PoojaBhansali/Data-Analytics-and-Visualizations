```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from regressors import stats


df = pd.read_xlsx('term_project_612.xlsx')
df

x = df.loc[:,['rating']].dropna()
y = df.loc[:,['salary_lo']].dropna()

xTrain, xTest, yTrain, yTest = train_test_split(x,y, test_size = 1/3)

model = LinearRegression().fit(xTrain, yTrain)

y_pred = model.predict(xTest)
print('predicted response:', y_pred, sep = '\n')

plt.scatter(xTest,yTest,color = 'blue')
plt.plot(xTrain,model.predict(xTrain),color = 'green')
plt.title('Min Salary vs. Rating')
plt.xlabel('Company Rating')
plt.ylabel('Company Min Salary')
plt.show

X = sm.add_constant(xTrain)
model = sm.OLS(yTrain, X).fit()
print(model.summary())
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import re
def salary_cate(row):
    if row['salary_expectation'] < 100000:
        return '0 < 100K'
    if row['salary_expectation'] < 150000:
        return '100K - 150K'
    return '> 150K'

def state_cate(row):
    if (
        'CA' in row['Location'] or 
        'San Francisco' in row['Location'] or 
        'California' in row['Location']
    ):
        return 'California'
    
    if 'NY' in row['Location'] or 'New York' in row['Location']:
        return 'New York'
    
    return 'Others'
    
df_job['Salary'] = df_job.apply(lambda row: salary_cate(row), axis=1)
df_job['State'] = df_job.apply(lambda row: state_cate(row), axis=1)

salary_state=pd.crosstab(df_job['Salary'],df_job['State'])
salary_state.plot(kind='bar')
plt.legend(title='State',loc='upper right')
plt.title('Expected Salary in Different Cities')
plt.show()
```

```python
from collections import Counter

skills_arr = []
for skill_row in df_job['Skills']:
    skills_arr.extend(skill_row.split(','))
    
skills_freq = Counter(skills_arr).most_common(4)

labellables = [s[0] for s in skills_freq]
sizes = [s[1] for s in skills_freq]
colors = ['#86d0f1', '#FD8D8C', '#a5d46e', '#f19464']
fig1, ax1 = plt.subplots(figsize=(5, 6)) 
skill_pie, _, _ = ax1.pie(
    sizes, 
    labels=labellables,
    colors=colors,
    autopct='%.2f%%',
    pctdistance = 0.7,   
    shadow=True, 
    startangle=90
)
ax1.legend(
    skill_pie, 
    labellables,
    title="Skills",
    bbox_to_anchor=(1, 1),
)
ax1.axis('equal')
ax1.set_title('Popular Skills for Data Scientists')
plt.show()
```

```python
#Let's take a look at the salary gap in each industry, there is not a big difference within industries.
lower_mean, lower_std = ((103160.25000,99173.67273,98559.08000,110980.49320,94265.54545),
(29880.94822,25245.59417,33918.11188,25060.53532,20814.86223))
higher_mean, higher_std = ((139432.0000,139357.6182,141513.3600,152513.4658,131797.0000),
(41917.71486,32586.0565,36399.69797,32882.39532,25779.46743))

ind = np.arange(len(lower_mean))
width = 0.35

fig,ax = plt.subplots()
rects1 = ax.bar(ind - width/2,lower_mean,width,yerr = lower_std,
               color = 'SkyBlue', label = 'lower bound')
rects2 = ax.bar(ind + width/2, higher_mean, width, yerr = higher_std,
               color = 'IndianRed',label = 'higher bound')

ax.set_ylabel('Salary range')
ax.set_title('Salary range by industry')
ax.set_xticks(ind)
ax.set_xticklabels(('education','finance','health','IT','others'))
ax.legend()

def autolabel(rects,xpos = 'center'):
    """
    Attach a text label above each bar in *rects*, displaying its height.
    
    *xpos* indicates which side to place the text w.r.t, the center of the bar.
    It can be one of the following {'center','right','left'}
    """
    xpos = xpos.lower()
    ha = {'center':'center','right':'left','left':'right'}
    offset = {'center':0.5,'right':0.57, 'left':0.43}
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()*offset[xpos], 1.01*height,
               '{}'.format(height), ha = ha[xpos], va = 'bottom', rotation = 'vertical')
autolabel(rects1,'left')
autolabel(rects2,'right')

plt.show()
```
